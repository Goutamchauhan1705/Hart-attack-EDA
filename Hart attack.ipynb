```python
cat_list = ['sex','exang', 'ca', 'cp', 'fbs', 'restecg', 'slope', 'thal']
con_list = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
fig = plt.figure(figsize=(15,15))
for i, col in enumerate(cat_list):
    axi = plt.subplot(2, 4, i+1)
    sns.countplot(ax=axi, data=data, x=col).set_title(col)
    axi.set_xlabel('')
    axi.set_ylabel('')
```

```python
fig = plt.figure(figsize=(8,15))
for i, col in enumerate(con_list):
    axi = plt.subplot(3, 3, i+1)
    sns.boxenplot(ax=axi, y=data[col]).set_title(col)
    axi.set_ylabel('')
```

```python
for i, col in enumerate(con_list[1::]):
    axi = sns.displot(data=data, x=col,hue='exang',  kde=True)
    
```

```python
fig = plt.figure(figsize=(5,5))
ax = sns.countplot(data=data, x='exang')
ax.set_xlabel('')
ax.set_xticklabels(['low proness', 'high proness'])
```
